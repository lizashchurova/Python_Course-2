```python
import vk
session = vk.Session(access_token='6f66ac206f66ac206f66ac20176f02423366f666f66ac203449c117ac780662fc9cefd1')
api = vk.API(session, v='5.131')
```

```python
friends_ids = api.friends.get(user_id=147830581)['items']
```

```python
friends = api.users.get(user_ids=friends_ids)
```

```python
walls = {}
for friend in friends:
    try:
        wall = api.wall.get(owner_id=friend['id'])
        walls[friend['id']] = wall
    except Exception as e:
        print(e)
```

```python
posts = []
for wall in walls.values():
#     print(wall)
    posts += wall['items']
```

```python
import pandas as pd

df = pd.DataFrame(posts)
df.head()
```

```python
df.drop(['id', 'date', 'post_type', 'attachments', 'views', 'donut', 'carousel_offset', 'edited', 'final_post', 'is_pinned', 'geo'], axis='columns', inplace=True)
df.head()
```

```python
comments_count = []
for i in range(df.shape[0]):
    comments_count.append(df.iloc[i]['comments']['count'])
df['comments_count'] = comments_count
```

```python
likes_count = []
for i in range(df.shape[0]):
    likes_count.append(df.iloc[i]['likes']['count'])
df['likes_count'] = likes_count
```

```python
df.drop(['comments', 'likes', 'reposts'], axis='columns', inplace=True)
df.head()
```
